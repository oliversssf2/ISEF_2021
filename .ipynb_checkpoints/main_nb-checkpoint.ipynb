```python
import matplotlib.pyplot as plt
import numpy as np
import dplm_base
```

```python
dplm_instance = dplm_base.dplm('para1.csv')
dplm_instance.show_dplm_config()
# dplm_instance.calculate_moment
```

```python
moment_spring, moment_weight, moment_total = [[dplm_instance.calculate_moment(0.1,0.3,0.2, item)[i] for item in range(-50,51)] for i in range(0,3)]
```

```python
plt.plot(range(-50,51), moment_spring, label = 'moment_spring')
plt.plot(range(-50, 51), moment_weight, label = 'moment_weight')
plt.plot(range(-50, 51), moment_total, label = 'moment_total')
plt.legend()
plt.show()
```

```python
moment_spring = [dplm_instance.calculate_moment(0.2,0.3,0.2, item)[0] for item in range(-50,51)]
moment_weight = [dplm_instance.calculate_moment(0.2,0.3,0.2, item)[1] for item in range(-50,51)]
moment_total = [dplm_instance.calculate_moment(0.2,0.3,0.2, item)[1] for item in range(-50,51)]
```

```python
plt.plot(range(-50,51), moment_spring, label = 'moment_spring')
plt.plot(range(-50, 51), moment_weight, label = 'moment_weight')
plt.plot(range(-50, 51), moment_total, label = 'moment_total')
plt.legend()
plt.show()
```
